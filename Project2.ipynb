# How many passengers were aboard the titanic in each class?
Below shows unfiltered data:

```python
import pandas as pd

pax = pd.read_csv('titanic.csv')
pax
```

Below shows results:

```python
pax = pax.groupby(['class'])['class'].count()
print(pax)
```
